```python
import geopandas as gpd
import geodatasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
weekdays = pd.read_csv('amsterdam_weekdays.csv')
weekends = pd.read_csv('amsterdam_weekends.csv')
```

```python
# processing weekdays file
gdf = gpd.GeoDataFrame(weekdays, geometry=gpd.points_from_xy(weekdays.lng, weekdays.lat))
crs = 'EPSG:4326'  # WGS84 coordinate reference system
weekdays_gdf = gpd.GeoDataFrame(gdf, crs=crs, geometry='geometry')
weekdays_gdf['log_realSum'] = np.log(weekdays_gdf['realSum'])
weekdays_gdf
```

```python
# processing weekends file
gdf = gpd.GeoDataFrame(weekends, geometry=gpd.points_from_xy(weekends.lng, weekends.lat))
crs = 'EPSG:4326'  # WGS84 coordinate reference system
weekends_gdf = gpd.GeoDataFrame(gdf, crs=crs, geometry='geometry')
weekends_gdf['log_realSum'] = np.log(weekends_gdf['realSum'])
weekends_gdf
```

```python
# folium map weekdays' log of real sum prices as a clorometh 

n = weekdays_gdf.explore(
    column = 'log_realSum',
    scheme = 'quantiles',
    tooltip = 'realSum',
    popup = ['room_type','person_capacity'])

folium.TileLayer("CartoDB positron", show = False).add_to(n)
folium.LayerControl().add_to(n)

n.save('amsterdam_weekdays_map.html')
n
```

```python
# folium map of weekends' log of real sum as a clorometh with no base

m = weekends_gdf.explore(
    column = 'log_realSum',
    scheme = 'quantiles',
    tooltip = 'realSum',
    name = 'Airbnb Prices in Amsterdam on Weekends',
    popup = ['room_type','person_capacity'])

folium.TileLayer("CartoDB positron", show = False).add_to(m)
folium.LayerControl().add_to(m)

m.save('amsterdam_weekends_map.html')
m
```
