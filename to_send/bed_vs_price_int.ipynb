```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import altair as alt
```

```python
weekdays = pd.read_csv('amsterdam_weekdays.csv')
weekdays['DofW'] = 'Weekday'
weekdays.head()
```

```python
weekends = pd.read_csv('amsterdam_weekends.csv')
weekends['DofW'] = 'Weekend'
weekends.head()
```

```python
combined_data = pd.concat([weekends, weekdays], ignore_index = True)
```

```python
selection = alt.selection_interval()

points = alt.Chart(combined_data).mark_point().encode(
    y=alt.X('realSum:Q', axis=alt.Axis(title='Price (EU)')),
    x=alt.Y('bedrooms:Q', axis=alt.Axis(title='# of Bedrooms')),
    color=alt.Color('DofW:N', title='Day of Week'),
    tooltip=[alt.Tooltip('dist:Q', title='Miles from City Center', format='.2f'),
        alt.Tooltip('realSum:Q', title='Price', format=',.2f'),
        alt.Tooltip('bedrooms:Q', title='Number of Bedrooms')]
).add_params(
    selection
).properties(
    width = 500,
    height = 300
)

points
```

```python
points.save('price_vs_num_beds.html')
```
