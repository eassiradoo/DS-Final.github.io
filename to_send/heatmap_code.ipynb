```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

amsterdam_weekend_data = pd.read_csv('amsterdam_weekends.csv')
amsterdam_weekday_data = pd.read_csv('amsterdam_weekdays.csv')

# Add a new column to each dataset to indicate the type of day
amsterdam_weekend_data['day_type'] = 0
amsterdam_weekday_data['day_type'] = 1

# Concatenate the datasets
df = pd.concat([amsterdam_weekend_data, amsterdam_weekday_data], ignore_index=True)
df = df.iloc[:, 1:]
```

```python
df.drop(columns=['lat', 'lng', 'attr_index','rest_index'], inplace=True)
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='magma', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of Amsterdam Airbnb Data')
plt.savefig('correlation_heatmap.png', bbox_inches='tight')  # Save the plot with title
plt.xticks(rotation=45, ha='right')
plt.show()
```
